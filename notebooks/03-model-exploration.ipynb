```python
# Standard batch of 2 RGB images sized 224x224
INPUT_SIZE = (2, 3, 224, 224)
```

## RESNET18 

```python
print("=================== RESNET18 ARCHITECTURE REPORT ===================")
resnet_model = create_model(model_name="resnet18",num_classes=1,freeze_backbone=True)
summary(
    resnet_model, 
    input_size=INPUT_SIZE,
    col_names=["input_size", "output_size", "num_params", "trainable"],
    depth=3
)
```
